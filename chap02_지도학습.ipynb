# 지도학습(supervised learning)
- 입력과 출력 샘플 데이터가 있고, 주어진 입력으로부터 출력을 예측하고자 할 때 사용한다.
- 보통 학습데이터를 만드는데 많은 노력이 필요하다. (ex) 보이스피싱인지 아닌지 라벨링
- 지도학습에는 분류(classification)와 회귀(regression)이 있다.
- 분류(classification)
    - **미리정의된, 가능성 있는 여러 레이블 클래스(class label)중 하나를 예측하는 것**
    - 이진분류(binary classification)(예,아니오, 보이스피싱)와 다중분류(multi classfication)(붓꽃품종)
    - 이진분류에서 한 클래스를 양성(positive)클래스, 다른 클래스를 음성(negative)클래스라고 한다.
    - 양성 클래스 : 학습하고자하는 대상 : ex)보이스 피싱(yes)
    - 출력값 사이에 연속성이 없다. ex)영어, 프랑스.. : 영어와 프랑스 사이의 언어 : 없음.
- 회귀(regression)
    - 연속적인 숫자(=부동소수점수=실수)를 예측하는 것
    - ex)어떤 사람의 교육수준, 나이, 주거지를 바탕으로 연간소득(또는 범위)를 예측하는 것

# 일반화, 과대적합, 과소적합
- 모델이 **처음 보는 새로운 데이터**에 대해 정확하게 예측할 수 있다면 이를 훈련세트에서 테스트세트로  
일반화(generalization)되었다고 한다.
- 가능한 한 정확하게 일반화되도록 해야한다.
- 과대적합(overfitting)
    - 모델이 훈련 세트의 각 샘플에 너무 가깝게 맞춰져서 새로운 데이터에 일반화되기 어려울 때 일어난다.
- 과소적합(underfitting)
    - 반대로 너무 간단한 모델이 선택되는 것, 데이터 면면과 다양성을 잡아내지 못하고 훈련세트에도 잘 맞지 못한다.  

- 복잡한 모델 : 너무 상세한 규칙 : 45세 이상이고 자녀가 셋 미만이며 이혼하지 않은 고객(많은 특성(열))
- 간단한 모델 : 50세 이상인 사람은 보트를 사려한다.

- 이상적인 모델 : 일반화 성능이 최대가 되는 최적점의 모델


- 데이터 셋에 **다양한** 데이터 포인트가 **많을수록** 과대적합 없이 더 복잡한 모델을 만들 수 있다.

# K-최근접 이웃
- k-NN(k-NearestNeighbors)알고리즘은 가장 간단한 머신러닝 알고리즘
- 훈련데이터셋을 그냥 저장하는 것이 모델을 만드는 과정의 전부.
- 새로운 데이터 포인트에 대해 예측할 땐 알고리즘이 훈련데이터셋에서 가장 가까운 포인트, 즉 '최근접 이웃'을 찾는다.

```python
mglearn.plots.plot_knn_classification(n_neighbors=1)
```

- 임의의 k개를 선택하면, 이웃이 더 많은 클래스를 레이블로 지정한다.

```python
mglearn.plots.plot_knn_classification(n_neighbors=3)
```

- 클래스가 여러 개 일 때도 각 클래스에 속한 이웃이 몇개인지를 헤아려 가장 많은 것 선택

```python
#일반화 성능 평가를 위해 데이터를 훈련세트와 테스트세트로 나눈다.
from sklearn.model_selection import train_test_split
X, y = mglearn.datasets.make_forge() #전체 세트 배당 

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=0)
```

```python
# KNeighborsClassifier를 임포트하고 객체를 만든다. (n = ~ 으로 캡슐화)
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3)
```

```python
#이제 훈련세트를 사용하여 모델을 학습시킨다.
#이때 KNeighborsClassifier에서의 학습은 예측할 때 이웃을 찾을 수 있도록 저장하는 것
clf.fit(X_train, y_train)
print(clf)
```

```python
#테스트 데이터에 대해 predict 메서드를 호출해서 예측한다.
#가장 가까운 이웃을 계산한다음 가장 많은 클래스를 찾는다.

print("테스트 예측", clf.predict(X_test))
```

```python
#모델이 얼마나 잘 일반화 되었는지 평가하기 위해 score 메서드에 테스트 데이터와 테스트 레이블을 넣어 호출
print("테스트 정확도 {:.2f}".format(clf.score(X_test, y_test)))
```

# KNeighborsClassifier 분석
- forge dataset은 열2개, 결과레이블 2개, 따라서 xy축에 열1,2가 들어가고 결과2개(도형2개)로 영역이 나뉨 
- 알고리즘이 클래스 0과 클래스 1로 지정한 영역으로 나뉘는 결정경계(decision boundary)로 볼 수 있다. 
- decision boundary는 열이 2개 이하인 경우 사용가능

```python
#fig = plt.figure(figsize=(5,10)) 공간생성 
#ax1 = fig.add_subplot(1,3,1) #fig라는 공간에 1행 3열의 첫번째 공간 만듬 

#위의 과정을 

fig, axes = plt.subplots(1,3, figsize=(10,3)) #10,3 size 공간에 1*3 구조 빈 그래프 3(ax3개)개 만듬 
#size 반환 및 각 그래프를 ndarray로 반환

for n_neighbors, ax in zip([1,3,9], axes): # 짝지어 튜플 형식으로 만들어 2개의 변수에 할당하는 zip구조
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf = clf.fit(X,y) #이번에는 그냥 분포(경계)를 보는거니 데이터를 나누지 않고 훈련시킴 
    #위의 두행은 clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X,y)
    #fit 메소드는 self 오브젝트를 리턴합니다.
    #그래서 객체 생성과 fit 메소드를 한줄에 쓸수 있습니다? 이해안됨..
    mglearn.plots.plot_2d_separator(clf,X,fill=True, eps = 0.5, ax =ax ,alpha=.4)
    mglearn.discrete_scatter(X[:,0],X[:,1],y,ax=ax)
    ax.set_title("{}이웃".format(n_neighbors))
    ax.set_xlabel("특성0")
    ax.set_ylabel("특성1")
axes[0].legend(loc=3) #첫번째 그래프에 범례설정
```

- 왼쪽 그림(k=1) : 결정 경계가 훈련 데이터에 가깝게 따라감 : 복잡한 모델링 : overfitting
- 이웃을 늘릴 수록 결정 경계는 더 부드러워진다 = 단순한 모델링
- 이웃을 적게 사용하면 모델의 복잡도가 높아진다.
- 이웃을 많이 사용하면 복잡도가 낮아진다.
- 훈련 데이터 전체 개수를 이웃의 수로 지정하는 극단적인 경우 : 모든 테스트 포인트가 같은 이웃(모든 훈련 데이터)를 가지게 되므로 테스트 포인트에 대한 예측은 모두 같은 값이된다. 즉 훈련 세트에서 가장 많은 데이터 포인트를 가진 클래스가 예측값이 된다. 

```python
#실제 유방암데이터로 모델의 복잡도와 일반화 사이의 관계를 입증할 수 있는지 살펴보자.
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer() #bunch 할당(keys, data, target, target_names, DESCR...)
print(cancer.target_names)

#전체 데이터의 양성과 악성의 비율을 살펴보자.
dict_1 = {n : v for n,v in zip(cancer.target_names, np.bincount(cancer.target))}
print("본 세트 비율(악성/양성) : {:.2f}".format(dict_1["malignant"]/dict_1['benign']))

X_train, X_test, y_train, y_test = train_test_split(\
 cancer.data, cancer.target, stratify = cancer.target, random_state = 66)

#stratify = cancer.target 인자의 의미는?
#y 값, 여기서는 양성/악성 그것의 비율을 전체세트, train세트, test세트에 동일하게 배분하겠다는 의미
#만약 본세트에서 양성과 악성의 비율이 0.6이라면 train, test세트에서 양성과 악성의 비율이 0.6임.
#훈련세트와 테스트 세트의 정확도를 비교하기 위해서 양성과 악성의 비율을 변인통제하기 위해서
#훈련세트세트의 비율과 테스트 세트의 비율이 같은 것과 다른것에서 테스트세트에서의 정확도를 비교하면
#당연 비율이 같은 것이 더 높은 확률로 정답을 맞출 수 있음.
#비율이 같은 것과 다른 것을 비교해보면 알 수 있음. 

#최적의 k 값을 찾기위해서 테스트 세트의 정확도를 비교해야하는데
#훈련 세트와 비율을 같게해서 좀더 명확하게 보는데 효과가 있다. 

#이를 확인하기 위한 코드

#훈련세트 양성 악성 비율 
dict_2 = {n : v for n,v in zip(cancer.target_names, np.bincount(y_train))}
print("훈련세트 비율(악성/양성) : {:.2f}".format(dict_2["malignant"]/dict_2['benign']))

#테스트 세트 양성 악성 비율 
dict_3 = {n : v for n,v in zip(cancer.target_names, np.bincount(y_test))}
print("테스트세트 비율(악성/양성) : {:.2f}".format(dict_3["malignant"]/dict_3['benign']))

#stratify인자가 없다면 이 비율들이 다름.

#random_state = 어떤 수 인자 의미
#해당 인자가 없다면 실행마다 데이터를 동일하게 분할하지 않음 : 매번 결과가 다름 
#숫자는 어떤 숫자를 입력해도 좋음 
#가령 0을 입력한것과 66을 입력한 것은 서로 출력결과가 다르지만 각각 결과는 유일하게 결정됨
#실제 상황에서는 거의 필요 없지만, 
#다른 사람들과 결과를 공유해야하거나 실험결과를 똑같이 재현해야 할 때는 유용하게 쓰인다.
# https://hyjykelly.tistory.com/44


#다시 본론으로 돌아와서 
training_accuracy = []
test_accuracy = []

#1에서 10까지 n_neighbors 를 적용 
neighbors_settings = range(1,11)

for n_neighbors in neighbors_settings:
    #모델 생성
    clf = KNeighborsClassifier(n_neighbors=n_neighbors) #캡슐화
    clf.fit(X_train, y_train) #trainset 저장(학습)
    #훈련 세트 정확도 저장  :: ?? 훈련세트에 대한 정확도 ?? 
    training_accuracy.append(clf.score(X_train, y_train))
    #일반화 정확도 저장
    test_accuracy.append(clf.score(X_test, y_test))

plt.plot(neighbors_settings, training_accuracy, label = "훈련 정확도")
plt.plot(neighbors_settings, test_accuracy, label = "테스트 정확도")
plt.ylabel("정확도")
plt.xlabel("n_neighbors")
plt.legend()
```

```python
# 이웃이 하나일 때 훈련 정확도가 1인 이유는?
```

- 이웃의 수가 하나일 때는 훈련 데이터에 대한 예측이 완벽하다.
- 하지만 이웃의 수가 늘어나면 모델은 단순해지고 훈련 데이터의 정확도는 줄어든다.
- 이웃을 하나 사용한 테스트 세트의 정확도는 이웃을 많이 사용했을 대 보다 낮다.
- 이것은 이웃이 하나인 모델은 너무 복잡한 모델임일 설명해준다.
- 반대로 이옷을 10개를 사용했을 때는 모델이 너무 단순해서 정확도는 더 나빠진다.
- 정확도가 중간 정도인 여섯개를 사용한 경우가 가장 좋다.

# k-최근접 이웃 회귀
- k-NN은 회귀 분석에도 쓰인다.
- KNeighborsRegressor의 weights 매개변수가 기본값 'uniform'일 때는 np.mean 함수를 사용하여 단순 평균을 계산
- distance일때는 거리를 고려한 가중치 평균(average)를 계산한다.


```python
mglearn.plots.plot_knn_regression(n_neighbors = 3)
```

```python
#scikit-learn에서 회귀를 이용한 k-최근접 이웃 알고리즘은 KNeighborsRegressor에 구현되어 있음 
from sklearn.neighbors import KNeighborsRegressor
X , y = mglearn.datasets.make_wave(n_samples=40)
print(y)

#wave 데이터셋을 훈련 세트와 테스트 세트로 나눈다.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

#이웃의 수를 3으로 하여 모델의 객체를 만든다.
reg = KNeighborsRegressor(n_neighbors=3)

#훈련데이터와 타깃을 사용하여 모델을 학습시킨다.
reg.fit(X_train, y_train)

#테스트 세트에 대해 예측
print("테스트 세트 예측:\n", reg.predict(X_test)) #본세트 40개에서 나눴으니 당연히 더 적음.
```

```python
#score메서드를 사용해 모델을 평가할 수 있다.
#이 메서드(score)은 회귀일 땐 R^2(결정 계수)를 반환한다.
#결정계수는 회귀 모델에서 예측의 적합도를 측정한 것으로 보통 0에서 1사이의 값이 된다.
#1은 예측이 완벽한 경우, 0은 훈련세트의 출력값인 y_train의 평균으로만 예측하는 모델임.
#또한 R^2은 음수가 될 수도 있는데, 이때는 예측과 타깃이 상반된 경향을 가지는 경우이다.
```

```python
print("테스트 세트 R^2 : {:.2f}".format(reg.score(X_test,y_test)))
```

```python
fig, axes = plt.subplots(1,3,figsize=(15,4))

#-3과 3 사이에 1000개의 데이터 포인트를 만든다.
line = np.linspace(-3,3,1000).reshape(-1,1) #reshape(-1,1)의미 https://rfriend.tistory.com/345
#만약 reshape(-1,1)을 하지 않는다면
for n_neighbors, ax in zip([1,3,9],axes):
    
    #1,3,9 이웃을 사용한 예측을 한다.
    
    #객체 생성
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    #훈련
    reg.fit(X_train, y_train)
    #
    ax.plot(line, reg.predict(line))
    ax.plot(X_train, y_train, '^', c = mglearn.cm2(0), markersize = 8) 
    ax.plot(X_test, y_test, 'v', c = mglearn.cm2(1), markersize = 8) #c = mglearn.cm2는 그냥 색정보
    #marker = '^'와 그냥 '^'차이 :: 이은 선 아님 점만 찍냐의 차이
    
    ax.set_title("{}이웃의 훈련 스코어 : {:.2f} 테스트 스코어: {:.2f}".format(\
                            n_neighbors, reg.score(X_train, y_train),reg.score(X_test, y_test)))
    ax.set_xlabel("특성")
    ax.set_ylabel("타깃")
    
    axes[0].legend(["모델 예측", "훈련 데이터/타깃", "테스트 데이터/타깃"], loc = 'best')
    #위의 ax 순서에 따라 배치됨.
```

- k-NN 회귀에서 이웃의 수를 1로 하면 훈련세트의 각 데이터 포인트가 예측에 주는 영향이 커서   
예측값이 훈련데이터 포인트를 모두 지나간다. (2개이상이면 평균을 하든 거리를 고려한 평균을 하든 하는데 1개면 그 값이니까)
- 이는 매우 불안정한 예측을 만들어 낸다. 이웃을 많이 사용하면 훈련데이터에는 잘 안 맞을 수 있지만 더 안정된 예측을 얻게된다.


- 일반적으로 KNeighbors 분류기에 중요한 매개변수는 2개다.
- 데이터 포인트 사이의 거리를 재는 방법과 이웃의 수.
- 실제 이웃의 수는 3개나 5개 정도로 적을 때 가장 잘 작동하지만, 이 매개변수는 잘 조정해야한다.
- 위에서 했던 방식으로 k = 1 ~ 10 정확도 그래프 그려서 확인하면 좋다.
- 거리재는 방법은 보통 잘 작동하는 유클리디안 거리방식 사용 
- k-NN의 장점은 이해하기 매우 쉬운 모델이고, 많이 조정하지 않아도 좋은 성능을 낸다.
- 단점은 훈련세트가 매우 크면(특성의 수나 샘플의 수가 클 경우) 예측이 느려진다.  
또한 k-NN 알고리즘을 사용할 땐 데이터를 전처리하는 과정이 중요, 특성 값 대부분이 0인 데이터 셋에는 잘 작동하지 않음, 현업에서는 잘 사용하지 않음
